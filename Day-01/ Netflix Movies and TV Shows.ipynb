```python
# Drop the null values in the DataFrame 
df.dropna(inplace = True)
```

```python
# Drop duplicates rows in the DtaFrame 
df_cleaned = df.drop_duplicates()
```

```python
text_columns = ['type', 'director', 'cast', 'country', 'rating', 'listed_in']
for col in text_columns:
    df_cleaned[col] = df_cleaned[col].astype(str).str.strip().str.title()
```

```python
df.head()
```

```python
# Convert 'date_added' to datetime format
df_cleaned['date_added'] = pd.to_datetime(df_cleaned['date_added'], errors='coerce')
df_cleaned.head()
```

```python
#Rename columns (lowercase and use underscores)
df_cleaned.columns = [col.strip().lower().replace(" ", "_") for col in df_cleaned.columns]
df_cleaned.head()
```

```python
# Fix data types
df_cleaned['release_year'] = df_cleaned['release_year'].astype(int)
df_cleaned.head()
```

```python
# Final check
print("\nData types after cleaning:\n", df_cleaned.dtypes)
```
